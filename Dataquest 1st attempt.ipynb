```python
# Select relevant features
features = df[['segment', 'beh_segment', 'active_ind']]
```

```python
# One-hot encoding for categorical features
features = pd.get_dummies(features)
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
# KNN clustering
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Set the number of clusters (k)
k = 5  # You can adjust this value based on your requirements

# Initialize the KNN model
knn = NearestNeighbors(n_neighbors=k)

# Fit the model and get the nearest neighbors
knn.fit(features_scaled)
distances, indices = knn.kneighbors(features_scaled)

# Calculate the cluster labels based on the nearest neighbors
cluster_labels = np.apply_along_axis(lambda v: np.argmax(np.bincount(v)), axis=1, arr=indices[:, 1:])
```

```python
# Add cluster labels to the original dataset
df['cluster'] = cluster_labels

# Print the cluster sizes
print(df['cluster'].value_counts())

# Explore the characteristics of each cluster
for cluster in df['cluster'].unique():
    print(f"Cluster {cluster}:")
    print(df[df['cluster'] == cluster][['segment', 'beh_segment', 'active_ind']].value_counts())
    print()
```

```python
# PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
features_pca = pca.fit_transform(features_scaled)

# t-SNE
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=42)
features_tsne = tsne.fit_transform(features_scaled)
```

```python
import matplotlib.pyplot as plt

# Scatter plot with PCA
plt.figure(figsize=(10, 6))
plt.scatter(features_pca[:, 0], features_pca[:, 1], c=cluster_labels, cmap='viridis')
plt.title('Clusters (PCA)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar()
plt.show()

# Scatter plot with t-SNE
plt.figure(figsize=(10, 6))
plt.scatter(features_tsne[:, 0], features_tsne[:, 1], c=cluster_labels, cmap='viridis')
plt.title('Clusters (t-SNE)')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.colorbar()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Prepare the data
X = df.drop(['is_cold_start', 'active_ind'], axis=1)
y = df['active_ind'].map({'Active': 1, 'Semi Active': 0, 'Inactive': 0})

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Train the model
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

# Evaluate the model
y_pred = rf.predict(X_test)
print("Overall Performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.3f}")
print(f"Precision: {precision_score(y_test, y_pred):.3f}")
print(f"Recall: {recall_score(y_test, y_pred):.3f}")
print(f"F1-score: {f1_score(y_test, y_pred):.3f}")
```

```python
# Filter cold start customers in the testing set
cold_start_test = X_test[X_test['is_cold_start']]
y_cold_start_test = y_test[X_test['is_cold_start']]

# Evaluate model performance on cold start customers
y_cold_start_pred = rf.predict(cold_start_test)
print("\nPerformance on Cold Start Customers:")
print(f"Accuracy: {accuracy_score(y_cold_start_test, y_cold_start_pred):.3f}")
print(f"Precision: {precision_score(y_cold_start_test, y_cold_start_pred):.3f}")
print(f"Recall: {recall_score(y_cold_start_test, y_cold_start_pred):.3f}")
print(f"F1-score: {f1_score(y_cold_start_test, y_cold_start_pred):.3f}")
```
